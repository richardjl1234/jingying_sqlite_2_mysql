## Example: SELECT Query

Both functions return a pandas DataFrame for SELECT queries.

```python
# Example SELECT query on SQLite
sqlite_result = sqlite_sql("SELECT * FROM payroll_details;")
print("SQLite Result:")
print(sqlite_result)
```

```python
# Example SELECT query on MySQL
mysql_result = mysql_sql("select count(*) FROM processes;")
mysql_result
```

## Example: UPDATE Query

UPDATE/INSERT/DELETE queries return `None`.

```python
# Example UPDATE query on SQLite
sqlite_sql("UPDATE employees SET salary = 5000 WHERE id = 1")
print("SQLite update completed")
```

```python
# Example UPDATE query on MySQL
mysql_sql("UPDATE employees SET salary = 6000 WHERE id = 1")
print("MySQL update completed")
```
